```python
import os
import pandas as pd
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision import models
import re
import string
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
import pickle
from scipy import sparse
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

randomseed = 42

# 1. 数据加载与预处理
print("=== 数据加载与预处理 ===\n")

class DataLoader:
    def __init__(self, data_dir='./data', train_label_file='train.txt'):
        self.data_dir = data_dir
        self.train_label_file = train_label_file
        
    def load_labels(self):
        labels_path = os.path.join(self.data_dir, self.train_label_file)
        df = pd.read_csv(labels_path)
        print(f"标签文件加载完成，共 {len(df)} 条数据")
        print(f"标签分布:\n{df['tag'].value_counts()}")
        return df
    
    def load_single_pair(self, guid):
        img_path = os.path.join(self.data_dir, f"{guid}.jpg")
        txt_path = os.path.join(self.data_dir, f"{guid}.txt")
            
        with open(txt_path, 'r', encoding='latin-1') as f:  #数据集使用'latin-1'编码
            text_content = f.read().strip()
            
        return img_path, text_content
    
    def create_dataset(self):
        labels_df = self.load_labels()
        data_records = []
        missing_files = []
        
        for _, row in labels_df.iterrows():
            guid = str(row['guid'])
            tag = row['tag']
            
            img_path, text = self.load_single_pair(guid)
            
            if img_path is None or text is None:
                missing_files.append(guid)
                continue
                
            data_records.append({
                'guid': guid,
                'image_path': img_path,
                'text': text,
                'tag': tag
            })
        
        dataset_df = pd.DataFrame(data_records)
        print(f"数据集创建完成，共 {len(dataset_df)} 条有效数据")
        print(f"缺失文件: {len(missing_files)} 个")
        return dataset_df

class DataSplitter:
    def __init__(self, val_size=0.1, random_state=42):
        self.val_size = val_size
        self.random_state = random_state
        
    def split_train_val(self, dataset_df, stratify=True):
        if stratify:
            train_df, val_df = train_test_split(
                dataset_df,
                test_size=self.val_size,
                random_state=self.random_state,
                stratify=dataset_df['tag']
            )
        else:
            train_df, val_df = train_test_split(
                dataset_df,
                test_size=self.val_size,
                random_state=self.random_state
            )
        
        print(f"训练集大小: {len(train_df)}")
        print(f"验证集大小: {len(val_df)}")
        
        print("\n训练集标签分布:")
        print(train_df['tag'].value_counts())
        
        print("\n验证集标签分布:")
        print(val_df['tag'].value_counts())
        
        return train_df, val_df

def save_split_datasets(train_df, val_df, save_dir='./data/splits'):
    os.makedirs(save_dir, exist_ok=True)
    
    train_path = os.path.join(save_dir, 'train_dataset.csv')
    train_df.to_csv(train_path, index=False)
    print(f"训练集已保存到: {train_path}")
    
    val_path = os.path.join(save_dir, 'val_dataset.csv')
    val_df.to_csv(val_path, index=False)
    print(f"验证集已保存到: {val_path}")

def load_split_datasets(save_dir='./data/splits'):
    train_path = os.path.join(save_dir, 'train_dataset.csv')
    val_path = os.path.join(save_dir, 'val_dataset.csv')
    
    train_df = pd.read_csv(train_path)
    val_df = pd.read_csv(val_path)
    
    print(f"训练集加载: {len(train_df)} 条")
    print(f"验证集加载: {len(val_df)} 条")
    return train_df, val_df

# 数据加载与预处理执行
print("1. 初始化数据加载器...")
data_loader = DataLoader()
    
print("\n2. 创建完整数据集...")
dataset_df = data_loader.create_dataset()
    
print("\n3. 划分训练集和验证集...")
splitter = DataSplitter(val_size=0.1, random_state=42)
train_df, val_df = splitter.split_train_val(dataset_df, stratify=True)
    
print("\n4. 保存划分结果...")
save_split_datasets(train_df, val_df, save_dir='./data/splits')
print("\n数据加载与预处理完成！")

# 2. 文本预处理与向量化
print("\n=== 文本预处理与向量化 ===")

# 下载NLTK数据
try:
    nltk.download('punkt')
    nltk.download('stopwords')
    nltk.download('wordnet')
except:
    print("NLTK数据已下载或下载失败，继续执行...")

# 文本数据预处理
class TextPreprocessor:
    def __init__(self, language='english'):
        self.language = language
        self.stop_words = set(stopwords.words(language))
        self.lemmatizer = WordNetLemmatizer()
        
        self.url_pattern = r'https?://\S+|www\.\S+'
        self.mention_pattern = r'@\w+'
        self.hashtag_pattern = r'#\w+'
    
    # 数据清洗 
    def clean_text(self, text):
        if not isinstance(text, str):
            return ""
        
        text = text.lower()
        text = re.sub(self.url_pattern, '', text)
        text = re.sub(self.mention_pattern, '', text)
        text = re.sub(self.hashtag_pattern, lambda x: x.group(0)[1:], text)
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s#@]', ' ', text)
        text = re.sub(r'\s+', ' ', text).strip()
        
        return text
    
    def tokenize(self, text):
        tokens = word_tokenize(text)
        return tokens
    
    def remove_stopwords(self, tokens):
        filtered_tokens = [word for word in tokens if word not in self.stop_words]
        return filtered_tokens
    
    def preprocess_pipeline(self, text):
        cleaned_text = self.clean_text(text)
        tokens = self.tokenize(cleaned_text)
        tokens = self.remove_stopwords(tokens)
        processed_text = ' '.join(tokens)
        return processed_text
    
    def analyze_text_data(self, texts):
        all_tokens = []
        original_lengths = []
        processed_lengths = []
        
        for text in texts:
            original_lengths.append(len(str(text)))
            processed = self.preprocess_pipeline(str(text))
            processed_lengths.append(len(processed))
            tokens = self.tokenize(processed)
            all_tokens.extend(tokens)
        
        vocab = set(all_tokens)
        token_counts = Counter(all_tokens)
        
        analysis = {
            'total_texts': len(texts),
            'avg_original_length': np.mean(original_lengths),
            'avg_processed_length': np.mean(processed_lengths),
            'vocab_size': len(vocab),
            'total_tokens': len(all_tokens),
            'avg_tokens_per_text': len(all_tokens) / len(texts),
            'most_common_tokens': token_counts.most_common(20)
        }
        return analysis

# 文本向量化
class TextVectorizer:
    def __init__(self, max_features=5000):
        self.max_features = max_features
        self.vectorizer = None
        self.is_fitted = False
        
    def fit(self, texts):
        self.vectorizer = TfidfVectorizer(
            max_features=self.max_features,
            ngram_range=(1, 2),
            min_df=2,
            max_df=0.95,
            stop_words='english'
        )
        self.vectorizer.fit(texts)
        self.is_fitted = True
        print(f"向量化器训练完成")
        print(f"特征维度: {len(self.vectorizer.get_feature_names_out())}")
    
    def transform(self, texts):
        return self.vectorizer.transform(texts)
    
    def save(self, filepath):     
        with open(filepath, 'wb') as f:
            pickle.dump({
                'max_features': self.max_features,
                'vectorizer': self.vectorizer,
                'is_fitted': self.is_fitted
            }, f)
        print(f"向量化器已保存到: {filepath}")

# 文本预处理与向量化执行
print("\n1. 加载数据集...")
train_df, val_df = load_split_datasets('./data/splits')
    
print("\n2. 初始化文本预处理器...")
preprocessor = TextPreprocessor(language='english')
    
print("\n3. 分析原始文本数据...")
analysis = preprocessor.analyze_text_data(train_df['text'])
print(f"原始文本分析结果:")
print(f"  总文本数: {analysis['total_texts']}")
print(f"  平均原始长度: {analysis['avg_original_length']:.1f} 字符")
print(f"  词汇表大小: {analysis['vocab_size']}")
print(f"  平均每文本词数: {analysis['avg_tokens_per_text']:.1f}")
    
print("\n4. 预处理文本数据...")
print("  处理训练集...")
train_texts_processed = train_df['text'].apply(
    lambda x: preprocessor.preprocess_pipeline(x)
)
    
print("  处理验证集...")
val_texts_processed = val_df['text'].apply(
    lambda x: preprocessor.preprocess_pipeline(x)
)

print("\n5. 保存预处理后的文本...")
train_df_processed = train_df.copy()
train_df_processed['text_processed'] = train_texts_processed
    
val_df_processed = val_df.copy()
val_df_processed['text_processed'] = val_texts_processed
    
train_df_processed.to_csv('./data/splits/train_processed.csv', index=False)
val_df_processed.to_csv('./data/splits/val_processed.csv', index=False)
print("预处理后的文本已保存")

print(f"\n=== 多模态情感分类 - 文本向量化 ===")
    
print("\n1. 加载预处理后的数据集...")
train_df = pd.read_csv('./data/splits/train_processed.csv')
val_df = pd.read_csv('./data/splits/val_processed.csv')
    
print("\n2. 初始化文本向量化器...")
vectorizer = TextVectorizer(max_features=2000)
    
print("\n3. 训练向量化器并转换训练集...")
train_texts = train_df['text_processed'].fillna('').tolist()
train_features = vectorizer.fit(train_texts)
train_features = vectorizer.transform(train_texts)
    
print("\n4. 转换验证集...")
val_texts = val_df['text_processed'].fillna('').tolist()
val_features = vectorizer.transform(val_texts)
    
print("\n5. 保存向量化器和特征...")
vectorizer.save('./data/splits/text_vectorizer.pkl')
    
sparse.save_npz('./data/splits/train_features.npz', sparse.csr_matrix(train_features))
sparse.save_npz('./data/splits/val_features.npz', sparse.csr_matrix(val_features))
print(f"\n文本向量化完成!")
print(f"向量化器已保存到: ./data/splits/text_vectorizer.pkl")

# 3. 图像处理与特征提取
print("\n=== 图像处理与特征提取 ===")

# 图像处理
class ImageProcessor:
    def __init__(self, image_size=(224, 224)):
        self.image_size = image_size
        self.transform = transforms.Compose([
            transforms.Resize(image_size),
            transforms.CenterCrop(image_size),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                               std=[0.229, 0.224, 0.225])
        ])
        
    def load_and_preprocess_image(self, image_path):
        image = Image.open(image_path).convert('RGB')
        image_tensor = self.transform(image)
        return image_tensor


# 图像特征提取
class ImageFeatureExtractor:
    # 使用resnet50模型进行特征提取
    def __init__(self, model_name='resnet50', device=None):
        self.model_name = model_name
        self.device = torch.device('cpu')
        
        self.model = self._load_pretrained_model(model_name)
        self.model.to(self.device)
        self.model.eval()
        
        self.feature_dim = self._get_feature_dimension()
        print(f"加载模型: {model_name}")
        print(f"特征维度: {self.feature_dim}")
    
    def _load_pretrained_model(self, model_name):
        model = models.resnet50(pretrained=True)
        model = torch.nn.Sequential(*list(model.children())[:-2])
        model.add_module('global_avg_pool', torch.nn.AdaptiveAvgPool2d((1, 1)))
        model.add_module('flatten', torch.nn.Flatten())
        
        for param in model.parameters():
            param.requires_grad = False
        
        return model
    
    def _get_feature_dimension(self):
        dummy_input = torch.randn(1, 3, 224, 224).to(self.device)
        with torch.no_grad():
            features = self.model(dummy_input)
        return features.shape[1]
    
    def extract_features_batch(self, image_tensors):
        if not image_tensors:
            return np.array([])
        
        batch_tensor = torch.stack(image_tensors).to(self.device)
        features_list = []
        batch_size = 32
        
        with torch.no_grad():
            for i in range(0, len(batch_tensor), batch_size):
                batch = batch_tensor[i:i+batch_size]
                features = self.model(batch)
                features_list.append(features.cpu().numpy())
        
        all_features = np.vstack(features_list)
        return all_features
    
    def extract_features_from_paths(self, image_paths, processor, batch_size=32):
        print(f"从 {len(image_paths)} 张图像提取特征...")
        
        features = []
        success_indices = []
        failed_indices = []
        
        for i in tqdm(range(0, len(image_paths), batch_size), desc="提取特征"):
            batch_paths = image_paths[i:i+batch_size]
            batch_tensors = []
            batch_success_indices = []
            
            for j, img_path in enumerate(batch_paths):
                img_tensor = processor.load_and_preprocess_image(img_path)
                if img_tensor is not None:
                    batch_tensors.append(img_tensor)
                    batch_success_indices.append(i + j)
                else:
                    failed_indices.append(i + j)
            
            if batch_tensors:
                batch_features = self.extract_features_batch(batch_tensors)
                features.append(batch_features)
                success_indices.extend(batch_success_indices)
        
        if features:
            all_features = np.vstack(features)
        else:
            all_features = np.array([])
        
        print(f"特征提取完成:")
        print(f"  成功: {len(success_indices)} 张图像")
        print(f"  失败: {len(failed_indices)} 张图像")
        
        return all_features, success_indices, failed_indices
    
    def save_features(self, features, filepath):
        np.save(filepath, features)
        print(f"特征已保存到: {filepath}")
        print(f"特征形状: {features.shape}")

# 图像处理与特征提取执行
print("\n1. 加载数据集...")
train_df = pd.read_csv('./data/splits/train_dataset.csv')
val_df = pd.read_csv('./data/splits/val_dataset.csv')
print(f"训练集: {len(train_df)} 条")
print(f"验证集: {len(val_df)} 条")
    
print("\n2. 初始化图像处理器...")
processor = ImageProcessor(image_size=(224, 224))
    
print("\n3. 初始化特征提取器...")
extractor = ImageFeatureExtractor(model_name='resnet50')
    
print("\n4. 提取训练集特征...")
train_image_paths = train_df['image_path'].tolist()
train_features, train_success_idx, train_failed_idx = extractor.extract_features_from_paths(
    train_image_paths, processor, batch_size=32
)
    
print("\n5. 提取验证集特征...")
val_image_paths = val_df['image_path'].tolist()
val_features, val_success_idx, val_failed_idx = extractor.extract_features_from_paths(
    val_image_paths, processor, batch_size=32
)

print("\n6. 保存特征...")
extractor.save_features(train_features, './data/splits/train_image_features.npy')
extractor.save_features(val_features, './data/splits/val_image_features.npy')
    
np.save('./data/splits/train_success_idx.npy', train_success_idx)
np.save('./data/splits/val_success_idx.npy', val_success_idx)

# 4. 多模态特征融合
print("\n=== 多模态特征融合 ===")

# 加载特征
class FeatureLoader:
    def __init__(self, data_dir='./data/splits'):
        self.data_dir = data_dir
        
    def load_text_features(self):
        train_text = sparse.load_npz(os.path.join(self.data_dir, 'train_features.npz')).toarray()
        val_text = sparse.load_npz(os.path.join(self.data_dir, 'val_features.npz')).toarray()
        print(f"文本特征形状:")
        print(f"  训练集: {train_text.shape}")
        print(f"  验证集: {val_text.shape}")
        return train_text, val_text
    
    def load_image_features(self):
        train_image = np.load(os.path.join(self.data_dir, 'train_image_features.npy'))
        val_image = np.load(os.path.join(self.data_dir, 'val_image_features.npy'))
        print(f"图像特征形状:")
        print(f"  训练集: {train_image.shape}")
        print(f"  验证集: {val_image.shape}")
        return train_image, val_image
    
    def load_labels(self):
        train_df = pd.read_csv(os.path.join(self.data_dir, 'train_dataset.csv'))
        val_df = pd.read_csv(os.path.join(self.data_dir, 'val_dataset.csv'))
        
        le = LabelEncoder()
        all_labels = pd.concat([train_df['tag'], val_df['tag']])
        le.fit(all_labels)
        
        train_labels = le.transform(train_df['tag'])
        val_labels = le.transform(val_df['tag'])
        
        print(f"标签分布:")
        print(f"  训练集: {dict(zip(*np.unique(train_labels, return_counts=True)))}")
        print(f"  验证集: {dict(zip(*np.unique(val_labels, return_counts=True)))}")
        return train_labels, val_labels, le

    def align_features(self, text_features, image_features):
        if len(text_features) != len(image_features):
            print(f"警告: 文本和图像特征数量不一致 ({len(text_features)} vs {len(image_features)})")
            common_size = min(len(text_features), len(image_features))
            text_features = text_features[:common_size]
            image_features = image_features[:common_size]
            print(f"截取前 {common_size} 个样本进行对齐")
        return text_features, image_features

# 多模态特征融合
class MultimodalFeatureFusion:
    '''
    超参数：
    text_weight： 文本权重
    image_weight： 图像权重
    '''
    def __init__(self):
        self.text_weight = 0.8
        self.image_weight = 0.2

    def weighted_concat(self, text_features, image_features, text_weight=0.5, image_weight=0.5):
        print(f"加权拼接融合: 文本权重{text_weight}, 图像权重{image_weight}")
        
        text_norm = text_features / (np.linalg.norm(text_features, axis=1, keepdims=True) + 1e-8)
        image_norm = image_features / (np.linalg.norm(image_features, axis=1, keepdims=True) + 1e-8)
        
        weighted_text = text_norm * text_weight
        weighted_image = image_norm * image_weight
        
        fused_features = np.concatenate([weighted_text, weighted_image], axis=1)
        print(f"融合后维度: {fused_features.shape[1]}")
        
        return fused_features
    
    def fuse(self, text_features, image_features):
        return self.weighted_concat(text_features, image_features, self.text_weight, self.image_weight)

class FeatureProcessor:
    def __init__(self):
        self.text_scaler = None
        self.image_scaler = None
        self.fused_scaler = None
        self.pca = None
    
    # 特征标准化
    def normalize_features(self, train_features, val_features, feature_type):
        
        if feature_type == 'text':
            if self.text_scaler is None:
                self.text_scaler = StandardScaler()
                train_normalized = self.text_scaler.fit_transform(train_features)
            else:
                train_normalized = self.text_scaler.transform(train_features)
            val_normalized = self.text_scaler.transform(val_features) if val_features is not None else None
        elif feature_type == 'image':
            if self.image_scaler is None:
                self.image_scaler = StandardScaler()
                train_normalized = self.image_scaler.fit_transform(train_features)
            else:
                train_normalized = self.image_scaler.transform(train_features)
            val_normalized = self.image_scaler.transform(val_features) if val_features is not None else None
        else:
            if self.fused_scaler is None:
                self.fused_scaler = StandardScaler()
                train_normalized = self.fused_scaler.fit_transform(train_features)
            else:
                train_normalized = self.fused_scaler.transform(train_features)
            val_normalized = self.fused_scaler.transform(val_features) if val_features is not None else None
        
        return train_normalized, val_normalized
    
    # PCA
    def pca_reduction(self, features, n_components=256, fit=True):
        from sklearn.decomposition import PCA
        
        if fit or self.pca is None:
            print(f"使用PCA降维: {features.shape[1]} -> {n_components} 维")
            self.pca = PCA(n_components=min(n_components, features.shape[1]), random_state=42)
            reduced_features = self.pca.fit_transform(features)
            explained_variance = np.sum(self.pca.explained_variance_ratio_)
            print(f"解释方差比: {explained_variance:.4f}")
        else:
            reduced_features = self.pca.transform(features)
        return reduced_features

# 多模态特征融合执行
print("\n1. 初始化特征加载器...")
loader = FeatureLoader(data_dir='./data/splits')
    
print("\n2. 加载文本特征...")
train_text, val_text = loader.load_text_features()
    
print("\n3. 加载图像特征...")
train_image, val_image = loader.load_image_features()
    
print("\n4. 加载标签...")
train_labels, val_labels, label_encoder = loader.load_labels()
    
print("\n5. 对齐特征...")
train_text, train_image = loader.align_features(train_text, train_image)
val_text, val_image = loader.align_features(val_text, val_image)

print("\n6. 特征标准化...")
processor = FeatureProcessor()
train_text_norm, val_text_norm = processor.normalize_features(train_text, val_text, feature_type='text')
train_image_norm, val_image_norm = processor.normalize_features(train_image, val_image, feature_type='image')
    
print("\n7. 特征融合...")
fusion = MultimodalFeatureFusion()
train_fused = fusion.fuse(train_text_norm, train_image_norm)
val_fused = fusion.fuse(val_text_norm, val_image_norm)
    
print("\n8. 标准化融合特征...")
train_fused_norm, val_fused_norm = processor.normalize_features(train_fused, val_fused, feature_type='fused')
    
print("\n9. 特征降维...")
if train_fused_norm.shape[1] > 512:
    print("应用PCA降维...")
    train_fused_reduced = processor.pca_reduction(train_fused_norm, n_components=512, fit=True)
    val_fused_reduced = processor.pca_reduction(val_fused_norm, n_components=512, fit=False)
    train_fused_norm = train_fused_reduced
    val_fused_norm = val_fused_reduced

print("\n10. 保存融合特征...")
save_dir = './data/splits/fusion'
os.makedirs(save_dir, exist_ok=True)

np.save(os.path.join(save_dir, 'train_fused.npy'), train_fused_norm)
np.save(os.path.join(save_dir, 'val_fused.npy'), val_fused_norm)
np.save(os.path.join(save_dir, 'train_text_norm.npy'), train_text_norm)
np.save(os.path.join(save_dir, 'val_text_norm.npy'), val_text_norm)
np.save(os.path.join(save_dir, 'train_image_norm.npy'), train_image_norm)
np.save(os.path.join(save_dir, 'val_image_norm.npy'), val_image_norm)
np.save(os.path.join(save_dir, 'train_labels.npy'), train_labels)
np.save(os.path.join(save_dir, 'val_labels.npy'), val_labels)

print(f"融合特征已保存到: {save_dir}")
print(f"融合特征形状:")
print(f"  训练集: {train_fused_norm.shape}")
print(f"  验证集: {val_fused_norm.shape}")
print("\n特征融合完成!")

# 5. 基线模型训练
print("\n=== 基线模型训练 ===")

class BaselineClassifier:
    def __init__(self, model_type='logistic', random_state=42):
        self.model_type = model_type
        self.random_state = random_state
        self.model = None
        
        if model_type == 'logistic':
            self.model = LogisticRegression(
                max_iter=1000,
                random_state=random_state,
                multi_class='multinomial',
                solver='lbfgs',
                class_weight='balanced'
            )
    
    def train(self, X_train, y_train, X_val, y_val):
        print(f"训练 {self.model_type} 模型...")
        print(f"训练集形状: {X_train.shape}, 标签分布: {np.bincount(y_train)}")
        
        self.model.fit(X_train, y_train)
        
        y_train_pred = self.model.predict(X_train)
        train_acc = accuracy_score(y_train, y_train_pred)
        train_f1 = f1_score(y_train, y_train_pred, average='weighted')
        
        y_val_pred = self.model.predict(X_val)
        val_acc = accuracy_score(y_val, y_val_pred)
        val_f1 = f1_score(y_val, y_val_pred, average='weighted')
        
        print(f"训练集准确率: {train_acc:.4f}")
        print(f"训练集F1分数: {train_f1:.4f}")
        print(f"验证集准确率: {val_acc:.4f}")
        print(f"验证集F1分数: {val_f1:.4f}")
        
        return {
            'train_accuracy': train_acc,
            'train_f1': train_f1,
            'val_accuracy': val_acc,
            'val_f1': val_f1
        }
    
    def predict(self, X):
        return self.model.predict(X)
    
    def save_model(self, filepath):
        joblib.dump(self.model, filepath)
        print(f"模型已保存到: {filepath}")
    
    def load_model(self, filepath):
        self.model = joblib.load(filepath)
        print(f"模型已从 {filepath} 加载")

def quick_train():
    """快速训练一个模型作为测试"""
    print("训练模型测试...")
    
    # 1. 加载数据
    print("1. 加载数据...")
    fusion_dir = './data/splits/fusion'
    train_fused = np.load(os.path.join(fusion_dir, 'train_fused.npy'))
    val_fused = np.load(os.path.join(fusion_dir, 'val_fused.npy'))
    train_labels = np.load(os.path.join(fusion_dir, 'train_labels.npy'))
    val_labels = np.load(os.path.join(fusion_dir, 'val_labels.npy'))
    
    print(f"训练集: {train_fused.shape}")
    print(f"验证集: {val_fused.shape}")
    
    # 2. 训练逻辑回归模型
    print("\n2. 训练逻辑回归模型...")
    classifier = LogisticRegression(
        max_iter=1000,
        random_state=42,
        multi_class='multinomial',
        solver='lbfgs',
        class_weight='balanced'
    )
    
    classifier.fit(train_fused, train_labels)
    
    # 3. 评估模型
    print("\n3. 评估模型...")
    train_pred = classifier.predict(train_fused)
    train_acc = accuracy_score(train_labels, train_pred)
    print(f"训练集准确率: {train_acc:.4f}")
    
    val_pred = classifier.predict(val_fused)
    val_acc = accuracy_score(val_labels, val_pred)
    val_f1 = f1_score(val_labels, val_pred, average='weighted')
    print(f"验证集准确率: {val_acc:.4f}")
    print(f"验证集F1分数: {val_f1:.4f}")
    
    # 4. 保存模型
    print("\n4. 保存模型...")
    os.makedirs('./data/models', exist_ok=True)
    model_path = './data/models/baseline_logistic_quick.pkl'
    joblib.dump(classifier, model_path)
    print(f"模型已保存到: {model_path}")
    
    # 5. 可视化混淆矩阵
    print("\n5. 可视化混淆矩阵...")
    
    plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
    plt.rcParams['axes.unicode_minus'] = False
    
    cm = confusion_matrix(val_labels, val_pred)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=['negative', 'neutral', 'positive'],
               yticklabels=['negative', 'neutral', 'positive'])
    plt.xlabel('预测标签')
    plt.ylabel('真实标签')
    plt.title('验证集混淆矩阵 (逻辑回归)')
    plt.savefig('./data/splits/quick_confusion_matrix.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    # 6. 特征重要性分析
    print("\n6. 特征重要性分析...")
    if hasattr(classifier, 'coef_'):
        print("模型系数形状:", classifier.coef_.shape)
        avg_importance = np.mean(np.abs(classifier.coef_), axis=0)
        top_indices = np.argsort(avg_importance)[::-1][:10]
        
        print("\n最重要的10个特征索引:")
        for i, idx in enumerate(top_indices):
            print(f"  特征 {idx}: 平均重要性 = {avg_importance[idx]:.4f}")
    
    print("\n测试训练完成!")
    return classifier, val_acc, val_f1

# 执行快速训练
print("开始模型训练测试...")
classifier, val_acc, val_f1 = quick_train()
print(f"最终验证集准确率: {val_acc:.4f}, F1分数: {val_f1:.4f}")
```
